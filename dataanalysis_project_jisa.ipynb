#OUTLIER DETECTION USING IQR METHOD AND VISUALISATION USING BOXPLOT

```python
def detect_outliers_iqr(data, column):                            #IQR METHOD
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    print(f"Outliers detected in '{column}':")
    print(outliers)
    return outliers
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_columns:
    print(f"\nAnalyzing column: {col}")
    detect_outliers_iqr(df, col)
```

#FEATURE ENGINEERING

```python
def age_category(age):
    if age < 30:
        return 'Young'
    elif age < 50:
        return 'Adult'
    else:
        return 'Senior'
```

```python
df['Age_Category'] = df['Age'].apply(age_category)
```

```python
def income_category(income):
    if income < 4000:
        return 'Low'
    elif income < 8000:
        return 'Medium'
    elif income < 12000:
        return 'High'
    else:
      return 'Very High'
```

```python
df['Income_Category'] = df['MonthlyIncome'].apply(income_category)
```

```python
df['RecentlyPromoted'] = df['YearsSinceLastPromotion'].apply(lambda x: 'Yes' if x == 0 else 'No')
```
